```python
class Node:
    def __init__(self, value=None, left=None, right=None, parent=None):
        self.value = value
        self.left = left
        self.right = right
        self.parent = parent
        self.color = 'r'
```

```python
class RBTree:
    def __init__(self, root=None):
        self.root = root
        
    def sucessor(self, node):
        if node.right is not None:
            return self.minimum(node.right)
        y = node.parent
        while y is not None and node == y.right:
            node = y
            y = y.parent
        return y
    
    def left_rotate(self, node):
        aux_node = node.right
        node.right = aux_node.left
        aux_node.left.parent = node
        aux_node.parent = node.parent
        
        if aux_node.parent is not None:
            self.root = aux_node
        elif node == node.parent.left:
            node.parent.left = aux_node
        else:
            node.parent.right = aux_node
        aux_node.left = node
        node.parent = aux_node
        
    def right_rotate(self, node):
        aux_node = node.left
        node.left = aux_node.right
        aux_node.right.parent = node
        aux_node.parent = node.parent
        
        if aux_node.parent is not None:
            self.root = aux_node
        elif node == node.parent.right:
            node.parent.right = aux_node
        else:
            node.parent.left = aux_node
        aux_node.right = node
        node.parent = aux_node
        
    
    def insert(self, node):
        y = None
        x = self.root
        while x is not None:
            y = x
            if node.value < x.value:
                x = x.left
            else:
                x = x.right
        node.parent = y
        if y is None:
            self.root = node
        elif node.value < y.value:
            y.left = node
        else:
            y.right = node
        
        node.left = None
        node.right = None
        node.color = 'r'
        self.insert_fix_up(node)
        
    def insert_fix_up(self, node):
        while node.parent.color == 'r':
            if node.parent == node.parent.parent.left:
                y = node.parent.parent.right
                # Caso 1
                if y.color == 'r':
                    y.parent.color = 'b'
                    node.parent.parent.color = 'r'
                    node = node.parent.parent
                elif node == node.parent.right:
                    # Caso 2
                    node = node.parent
                    self.left_rotate(node)
                    # Caso 3
                    node.parent.color = 'b'
                    node.parent.parent.color = 'r'
                    self.right_rotate(node.parent.parent)
            else:
                y = node.parent.parent.left
                # Caso 1
                if y.color == 'r':
                    y.parent.color = 'b'
                    node.parent.parent.color = 'r'
                    node = node.parent.parent
                elif node == node.parent.left:
                    # Caso 2
                    node = node.parent
                    self.left_rotate(node)
                    # Caso 3
                    node.parent.color = 'b'
                    node.parent.parent.color = 'r'
                    self.right_rotate(node.parent.parent)
        self.root.color = 'b'
        
    def delete(self, node):
        if node.left is None or node.right is None:
            y = node
        else:
            y = self.sucessor(node)
        
        if y.left is not None:
            x = y.left
        else:
            x = y.right
        
        x.parent = y.parent
        
        if y.parent is None:
            self.root = x
        elif y == y.parent.left:
            y.parent.left = x
        else:
            y.parent.right = x
        
        if y != node:
            node.value = y.value
            
        if y.color == 'b':
            self.delete_fix_up(node)
        return y
    
    
    def delete_fix_up(self, node):
        while node != self.root and node.color == 'b':
            if node == node.parent.left:
                w = node.parent.right
                
                # Caso 1
                if w.color == 'r':
                    node.parent.color = 'b'
                    self.left_rotate(node.parent)
                    w = node.parent.right
                    
                
                if w.left.color == 'b' or w.right.color == 'b':
                    # Caso 2
                    w.color = 'r'
                    node = node.parent
                elif w.right.color == 'b':
                    # Caso 3
                    w.left.color = 'b'
                    w.color = 'r'
                    self.right_rotate(w)
                    w = node.parent.right
                    w.color = node.parent.color
                    node.parent.color = 'b'
                    w.right.color = 'b'
                    self.left_rotate(node.parent)
                    x = self.root
            else:
                w = node.parent.left
                
                # Caso 1
                if w.color == 'r':
                    node.parent.color = 'b'
                    self.left_rotate(node.parent)
                    w = node.parent.left
                    
                
                if w.left.color == 'b' or w.right.color == 'b':
                    # Caso 2
                    w.color = 'r'
                    node = node.parent
                elif w.left.color == 'b':
                    # Caso 3
                    w.right.color = 'b'
                    w.color = 'r'
                    self.right_rotate(w)
                    w = node.parent.left
                    w.color = node.parent.color
                    node.parent.color = 'b'
                    w.left.color = 'b'
                    self.left_rotate(node.parent)
                    x = self.root
        node.color = 'b'
        
    
```
